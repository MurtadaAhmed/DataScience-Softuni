```python
tb_tidy_no_missing.age.unique() # checking the unique values of the age column
```

```python
def process_age_group(age_group):
    ages = {"04": "0-4", "65": "65+", "u": "unknown"}
    if age_group in ages:
        return ages[age_group]
    elif len(age_group) == 3: # meaning the age group has 3 symbols >> ex: 014 >>> should return: 0-14
        return f"{age_group[0]}-{age_group[1:]}"
    else: # meaning the age group has 4 symbols >> ex: 1524 >>> should return: 15-24
        return f"{age_group[:2]}-{age_group[2:]}"
tb_tidy_no_missing["age"] = tb_tidy_no_missing.age.apply(process_age_group)
```

```python
tb_tidy_no_missing
```

```python
tb_tidy_no_missing.age.unique() # checking the unique values of the age column
```

```python
tb_tidy_no_missing.sex = tb_tidy_no_missing.sex.astype("category") # more memory efficient
tb_tidy_no_missing.age = tb_tidy_no_missing.age.astype("category") # more memory efficient
```

```python
tb_tidy_no_missing.dtypes
```

```python
tb_tidy_no_missing = tb_tidy_no_missing[["iso2", "year", "sex", "age", "cases"]] # reordering the columns
```

```python
tb_tidy_no_missing
```

```python
tb_tidy_no_missing = tb_tidy_no_missing.sort_values(["iso2", "year"]) # sorting the data by country and then by year
```

```python
tb_tidy_no_missing
```

```python
tb_tidy_no_missing = tb_tidy_no_missing.reset_index(drop=True) # resetting the index to start from 0
```

```python
tb_tidy_no_missing
```

```python
tb_tidy_no_missing.cases = tb_tidy_no_missing.cases.astype(int) # converting the cases column to integer
```

```python
tb_tidy_no_missing
```

```python
tb_tidy_no_missing.to_csv("demo_data/tb_tidy.csv", index=None) # saving the data to a csv file
```

```python
weather_data = pd.read_csv("demo_data/weather.csv")
```

```python
weather_data = weather_data.melt(id_vars=["id", "year", "month", "element"], var_name="day")
```

```python
weather_data
```

```python
weather_data.day = weather_data.day.str.slice(1).astype(int) # extracting the day and converting it to integer
```

```python
weather_data
```

```python
weather_data = weather_data.dropna() # dropping rows with NaN values
```

```python
weather_data
```

```python
weather_data.columns
```

```python
weather_data = weather_data.pivot_table(index=["id", "year", "month", "day"], columns="element", values="value") # converting the tmax and tmin rows to columns
```

```python
weather_data.columns
```

```python
weather_data = weather_data.reset_index()
```

```python
weather_data.columns
```

```python
weather_data["date"] = pd.to_datetime(weather_data[["year", "month", "day"]]) # creating a date column
```

```python
weather_data.columns
```

```python
weather_data
```

```python
weather_data = weather_data.drop(columns=["year", "month", "day"]) # deleting the year, month and day columns
```

```python
weather_data
```

```python
weather_data = weather_data[["id", "date", "tmax", "tmin"]] # reordering the columns
```

```python
weather_data
```

```python
weather_data.to_csv("demo_data/weather_tidy.csv", index=None) # saving the data to a csv file
```
